```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

# FEATURE ENGINEERING

```python
df = pd.read_parquet('data/generate_sample.parquet')
df
```

```python
df['hour'] = df.time_stamp.dt.hour
df['weekday_number'] = df.time_stamp.dt.weekday
df['weekday_name'] = df.time_stamp.dt.day_name()
df
```

```python
df = df.sort_values(by=['txn_maker_address', 'time_stamp'])
df['txn_time_gap'] = df.groupby('txn_maker_address')['time_stamp'].diff().dt.total_seconds()
df
```

```python
df['txn_value_to_token_price'] = df['txn_value']*100/df['token_price']
df['txn_value_to_liquidity'] = df['txn_value']*100/df['liquidity']
df['txn_value_to_market_cap'] = df['txn_value']*100/df['market_cap']
```

```python
df.to_parquet('data/historical_data.parquet', index=False)
```

```python
df = pd.read_parquet('data/historical_data.parquet')
df = df.sort_values(by=['txn_maker_address', 'time_stamp'])
df = df.drop_duplicates(subset=['txn_maker_address'], keep='last')
df
```

```python
df.to_parquet('data/most_current_data.parquet', index=False)
```

# EDA

```python
df['txn_value'].hist()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['time_stamp'], df['txn_value'], marker='o', linestyle='-')
plt.title('Value Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
df.set_index('time_stamp', inplace=True)
df_resampled = df.resample('1H').mean()  # Resample every 1 hour
plt.figure(figsize=(10, 6))
plt.plot(df_resampled.index, df_resampled['txn_value'], marker='o', linestyle='-')
plt.title('Mean Value Over 1-Hour Intervals')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df = df.reset_index(drop=True)
df_rapid_txn = df[df.txn_time_gap < 100]
```

```python
df_rapid_txn['txn_time_gap'].describe()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_rapid_txn['time_stamp'], df_rapid_txn['txn_time_gap'], marker='o', linestyle='-')
plt.title('Value Over Time')
plt.xlabel('Timestamp')
plt.ylabel('txn_time_gap')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.show()
```
